# Process data

```python
import json

with open('context_articles.json', 'r') as f:
  context_articles = json.load(f)

with open('iirc_train_dev/train.json', 'r') as f:
  train_json = json.load(f)

with open('iirc_train_dev/dev.json', 'r') as f:
  dev_json = json.load(f)
```

```python
import pandas as pd
from bs4 import BeautifulSoup
import spacy

nlp = spacy.blank("en")
nlp.add_pipe('sentencizer')

def clean_html(html):
  soup = BeautifulSoup(html, "html.parser")
  for data in soup(['style', 'script']):
      # Remove tags
      data.decompose()

  # return data by retrieving the tag content
  return ' '.join(soup.stripped_strings)

# df = pd.DataFrame(columns=['article', 'paragraph'])
jsonl = []
stride = 3
length = 3
for name, article in context_articles.items():
  article = clean_html(article)
  doc = nlp(article)
  sentences = [str(sent).strip() for sent in doc.sents]
  for i in range(0, len(sentences), stride):
    segment = ' '.join(sentences[i:i + length])
    aux_dict = {}
    aux_dict['id'] = f'{name}_{i}_{i + length}'
    aux_dict['contents'] = segment
    jsonl.append(aux_dict)
  # df_aux = pd.DataFrame(pars, columns=['paragraph'])
  # df_aux['article'] = name
  # df = pd.concat([df, df_aux],axis=0)
```

```python
with open('output/output.jsonl', 'w') as outfile:
    for entry in jsonl:
        json.dump(entry, outfile)
        outfile.write('\n')
```

```python
def get_context(jsonl, id):
  for item in jsonl:
    if item['id'] == id:
      return item['contents']
```

```python
import ast
from pyserini.index import IndexReader
total_questions = 0
total_contexts = 0
context_was_found = 0
context_was_found_5 = 0
none_answer = 0
number_answer = 0
span_answer = 0
binary_answer = 0
bad_answer = 0
recall = 0
removed_links = 0
index_reader = IndexReader('indexes/collection_jsonl/')
for passage in train_json:
    passage_links = [link_dict['target'].lower() for link_dict in passage['links']]

    for question in passage['questions']:
      total_questions += 1
      question_str = question['question']
      answer_info = question["answer"]
      a_type = answer_info["type"]
      if a_type == "span":
          span_answer += 1
      elif a_type == "value":
          number_answer += 1
          continue
      elif a_type == "binary":
          binary_answer += 1
          continue
      elif a_type == "none":
          none_answer += 1
          continue
      elif a_type == "bad":
          bad_answer += 1
          continue

      docids = []
      docids_scores = []
      for article_wdw in jsonl:
        key = article_wdw['id'].split('_')[0]
        if key in passage_links:
          docids.append(article_wdw['id'])
      
      for i in range(0, len(docids)):
        score = index_reader.compute_query_document_score(docids[i], question_str)
        docids_scores.append((docids[i], score))

      top_n = sorted(docids_scores, key = lambda x: x[1], reverse = True)[:500]

      gold_links = set([span["passage"] for span in question["context"] if span["passage"] != "main"])
      gold_contexts = [span["text"] for span in question["context"] if span["passage"] != "main"]

      for i, hit in enumerate(top_n):
        hit_context = get_context(jsonl,hit[0])

        for gold_context_str in gold_contexts:
          total_contexts += 1
          if gold_context_str in hit_context:
            context_was_found +=1
            recall += 1/len(gold_contexts)
```
